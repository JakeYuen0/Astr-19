```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

def exp_function(x):
    return np.exp(x)

y = exp_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = exp(x)", color='blue')
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Plot of Exponential Function")
plt.legend()

plt.savefig('exponential_plot.pdf', format='pdf')

plt.show()
```
