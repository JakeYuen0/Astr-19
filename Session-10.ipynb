```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rayleigh(scale = 1.0, size = 1000)

plt.figure(figsize = (10, 6))
plt.hist(data, bins = 100, color = 'blue', edgecolor = 'black')
           
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Histogram of 1000 Random Rayleigh Distribution Numbers')
           
plt.savefig("rayleigh-histopdf", bbox_inches = "tight", facecolor = 'white')
plt.show()
```
