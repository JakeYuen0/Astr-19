```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 1, 1000)

plt.figure(figsize = (10, 4))
plt.hist(data, bins = 100, color = 'blue', edgecolor = 'black')


plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Histogram of 1000 Uniformly Distributed Random Numbers')


plt.savefig("histogram.pdf", bbox_inches = "tight", facecolor = "white")
plt.show()
```
